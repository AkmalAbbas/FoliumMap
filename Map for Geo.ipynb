```python
import folium
from folium import FeatureGroup
```

```python
m = folium.Map(
    [38.58,-99.09],
    tiles='Stamen Terrain',
    zoom_start=6)
fg = FeatureGroup(name='Some icons')
```

```python
import pandas
```

```python
volcanoes = pandas.read_csv('Volcanoes.txt')
```

```python
lat,lon = list(volcanoes['LAT']),list(volcanoes['LON'])
```

```python
elev = list(volcanoes['ELEV'])
name = list(volcanoes['NAME'])
```

```python
html = """<h4>Volcano information:</h4>
Height: %s m<br>
Name: %s
"""

fg = FeatureGroup(name='Some icons')

def color_of_volcanoes(elevation):
    if elevation < 1000:
        return 'green'
    elif 1000 <= elevation < 3000:
        return 'orange'
    else:
        return 'red'
```

```python
for lt,ln,el,nm in zip(lat,lon,elev,name):
    iframe = folium.IFrame(html=html % (el,nm), width=200, height=100)
    fg.add_child(folium.CircleMarker(location=[lt,ln],radius = 6,
                                     popup=folium.Popup(iframe),
                                     fill_color = color_of_volcanoes(el),color='grey',fill_opacity=0.7))
```

```python
fg.add_child(folium.GeoJson(data = open('custom.geo.json').read(),
                            style_function = lambda x: {'fillColor':'Yellow'}))
```

```python
m.add_child(fg)
```

```python
m.save("Map.html")
```
